```python
import matplotlib.pyplot as plt
import numpy as np

# Função XOR
def xor_function(x, y):
    return x ^ y

# Criar dados para o plot
x_values = np.array([0, 0, 1, 1])
y_values = np.array([0, 1, 0, 1])
result = np.array([xor_function(x, y) for x, y in zip(x_values, y_values)])

# Plotar a função XOR
plt.figure(figsize=(3, 3))
#plt.scatter(x_values, y_values, c=result, cmap='hot', s=100)
# same but other colors
plt.scatter(x_values, y_values, c=result, cmap='cool', s=500)
# label for each color
#plt.colorbar(ticks=[0, 1])
# label for each point
for x, y, r in zip(x_values, y_values, result):
    plt.text(x, y, r, fontsize=15, ha='center', va='center')
# put some margin around the points so they aren't on the axes
plt.margins(0.2)

    
    
plt.title('Função XOR')
plt.xlabel('Entrada 1')
plt.ylabel('Entrada 2')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Função OR
def or_function(x, y):
    return x or y

# Função AND
def and_function(x, y):
    return x and y

# Função NOT
def not_function(x):
    return not x

# Função para criar os plots
def create_plot(function, title):
    # Criar dados para o plot
    x_values = np.array([0, 0, 1, 1])
    y_values = np.array([0, 1, 0, 1])
    result = np.array([function(x, y) for x, y in zip(x_values, y_values)])

    # Plotar a função
    plt.figure(figsize=(3, 3))
    plt.scatter(x_values, y_values, c=result, cmap='cool', s=500)
    for x, y, r in zip(x_values, y_values, result):
        plt.text(x, y, r, fontsize=15, ha='center', va='center')
    plt.margins(0.2)

    plt.title(title)
    plt.xlabel('Entrada 1')
    plt.ylabel('Entrada 2')
    plt.show()

# Criar plots para OR, AND e NOT
create_plot(or_function, 'Função OR')


```

```python
create_plot(and_function, 'Função AND')
```

```python
create_plot(lambda x, y: int(not not_function(x)), 'Função NOT')  # A função NOT tem apenas uma entrada
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Função XOR
def xor_function(x, y):
    return x ^ y

# Criar dados para o plot
x_values = np.array([0, 0, 1, 1])
y_values = np.array([0, 1, 0, 1])
result = np.array([xor_function(x, y) for x, y in zip(x_values, y_values)])

# Escolher a paleta de cores 'hot'
colors = plt.cm.hot(result)

# Plotar a função XOR com legenda
plt.scatter(x_values, y_values, c=colors, s=100, label='Resultado XOR')
plt.title('Função XOR')
plt.xlabel('Entrada 1')
plt.ylabel('Entrada 2')

# Adicionar legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```
